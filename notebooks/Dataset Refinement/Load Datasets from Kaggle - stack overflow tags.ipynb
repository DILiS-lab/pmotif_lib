```python
p = "/home/timgarrels/Downloads/kaggle_so_tags/"  # https://www.kaggle.com/datasets/stackoverflow/stack-overflow-tag-network
```

```python
import pandas as pd

nodes_df = pd.read_csv(p + "stack_network_nodes.csv")
nodes_df.head()
```

```python
node_name_lookup = {name: (i + 1) for i, name in enumerate(nodes_df["name"])}
node_index_lookup = {index: name for name, index in node_name_lookup.items()}
```

```python
import pandas as pd

edges_df = pd.read_csv(p + "stack_network_links.csv")
edges_df.head()
```

```python
import networkx as nx

g = nx.Graph()

for row in edges_df.iterrows():
    g.add_edge(
        node_name_lookup[row[1]["source"]],
        node_name_lookup[row[1]["target"]],
    )
```

```python
largest = set()
for component in nx.connected_components(g):
    if len(component) > len(largest):
        largest = component
        
g = nx.induced_subgraph(g, largest)
```

```python
nx.draw(g, labels={k: v for k, v in node_index_lookup.items() if k in g.nodes})
```
