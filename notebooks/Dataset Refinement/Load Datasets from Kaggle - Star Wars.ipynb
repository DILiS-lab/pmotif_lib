```python
p = "/home/timgarrels/Downloads/kaggle_starwars/"  # https://www.kaggle.com/datasets/ruchi798/star-wars?resource=download
```

```python
import json

with open(p + f"starwars-episode-{EPISODE}-mentions.json", "r") as f:
    data = json.load(f)
```

```python
import networkx as nx
```

```python
g = nx.Graph()


for edge in data["links"]:
    src = edge["source"] + 1
    dst = edge["target"] + 1
    g.add_node(edge["source"] + 1, data=src)
    g.add_node(edge["target"] + 1, data=dst)

    g.add_edge(edge["source"] + 1, edge["target"] + 1)


    
```

```python
node_names = {i + 1: n["name"] for i, n in enumerate(data["nodes"])}
node_names
```

```python
nx.draw(g, labels=node_names)
```
